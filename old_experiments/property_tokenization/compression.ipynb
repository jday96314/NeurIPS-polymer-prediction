# Try compressing morgan fingerprints

```python
train_poly_one_df = poly_one_df.head(n=len(poly_one_df)-10_000)
test_poly_one_df = poly_one_df.head(n=10_000)
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, MACCSkeys, rdFingerprintGenerator
import numpy as np
from tqdm import tqdm

morgan_fingerprint_generator = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
atom_pair_fingerprint_generator = rdFingerprintGenerator.GetAtomPairGenerator(fpSize=2048)

def get_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    morgan_fp = morgan_fingerprint_generator.GetFingerprint(mol)
    atom_pair_fp = atom_pair_fingerprint_generator.GetFingerprint(mol)
    maccs_fp = MACCSkeys.GenMACCSKeys(mol)
    combined_fp = np.concatenate([
        np.array(morgan_fp, dtype=np.uint8),
        np.array(atom_pair_fp, dtype=np.uint8),
        np.array(maccs_fp, dtype=np.uint8),
    ])

    return combined_fp

train_morgan_fingerprints = np.array([
    get_fingerprint(smiles)
    for smiles in tqdm(train_poly_one_df['smiles'])
])
poly_one_test_morgan_fingerprints = np.array([
    get_fingerprint(smiles)
    for smiles in tqdm(test_poly_one_df['smiles'])
])
host_morgan_fingerprints = np.array([
    get_fingerprint(smiles)
    for smiles in tqdm(host_df['SMILES'])
])
```

```python
from sklearn.decomposition import PCA
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, explained_variance_score
from sklearn.preprocessing import StandardScaler


def get_explained_variance(pca: PCA, validation_features: np.ndarray) -> float:
    projected               = pca.transform(validation_features)
    projected_variances     = projected.var(axis=0, ddof=0).sum()
    original_variances      = validation_features.var(axis=0, ddof=0).sum()
    fraction_variance_kept  = projected_variances / original_variances
    return fraction_variance_kept

def get_information_loss_metrics(train_fingerprints, poly_one_test_fingerprints, host_fingerprints, dimension_count):
    scaler = StandardScaler()
    normalized_train_fingerprints = scaler.fit_transform(train_fingerprints)
    normalized_poly_one_test_fingerprints = scaler.transform(poly_one_test_fingerprints)
    normalized_host_fingerprints = scaler.transform(host_fingerprints)

    pca = PCA(n_components=dimension_count)
    pca.fit(normalized_train_fingerprints)
    # pca.fit(host_fingerprints[:4000])

    poly_one_explained_variance = get_explained_variance(pca, normalized_poly_one_test_fingerprints)
    host_explained_variance = get_explained_variance(pca, normalized_host_fingerprints)
    # host_explained_variance = get_explained_variance(pca, host_fingerprints[4000:])

    return poly_one_explained_variance, host_explained_variance

dimension_counts = [2**exponent for exponent in range(10)]
poly_one_explained_variance_scores = []
host_explained_variance_scores = []
for dimension_count in tqdm(dimension_counts):
    poly_one_explained_variance, host_explained_variance = get_information_loss_metrics(train_morgan_fingerprints, poly_one_test_morgan_fingerprints, host_morgan_fingerprints, dimension_count)
    poly_one_explained_variance_scores.append(poly_one_explained_variance)
    host_explained_variance_scores.append(host_explained_variance)
```

```python
from matplotlib import pyplot as plt

plt.plot(dimension_counts, poly_one_explained_variance_scores, label="polyOne")
plt.plot(dimension_counts, host_explained_variance_scores, label="competition")
plt.ylim(0,1)

plt.title('Morgan Fingerprint Explained Variance')
plt.xlabel('# Dimensions')
plt.ylabel('Explained variance score')

plt.legend()

plt.show()
```
